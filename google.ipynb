## Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

## Load Dataset

```python
df = pd.read_csv(r"C:\Users\dell\Downloads\googleplaystore.csv")
df.head()
```

## Data Cleaning

```python
df = df[df['Rating'].notna()]
df = df[df['Rating'] <= 5]
```

```python
df['Rating'].isnull().sum()
```

```python
df['Reviews'] = df['Reviews'].astype(int)
```

```python
df['Installs'] = (
    df['Installs']
    .astype(str)
    .str.replace('[+,]', '', regex=True)
)

df['Installs'] = pd.to_numeric(df['Installs'], errors='coerce')
```

```python
df['Price'] = (df['Price'].astype(str).str.replace('$', '', regex=True).astype(float))
```

```python
def convert_size(size):
    if 'M' in str(size):
        return float(size.replace('M', ''))
    elif 'k' in str(size):
        return float(size.replace('k', '')) / 1024
    else:
        return np.nan

df['Size_MB'] = df['Size'].apply(convert_size)
```

```python
df['Revenue'] = df['Price'] * df['Installs']

df['Last Updated'] = pd.to_datetime(df['Last Updated'], errors='coerce')
df['Year'] = df['Last Updated'].dt.year
```

```python
df['Category'].value_counts().head(10).plot(kind='bar')
plt.title("Top App Categories")
plt.show()
```

```python
df.to_csv("googleplaystore_cleaned.csv", index=False)
```
